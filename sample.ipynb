```python
from faker import Faker
import pandas as pd
import numpy as np
import random
fake = Faker()
```

```python
files= 'data/data_500.csv'
df_csv = pd.read_csv(files)
df_csv.info()
df_csv.describe()
```

```python
df_csv.head()
```

```python
def generate_data_user(df_csv, num_user):
    user_ids = [fake.unique.uuid4() for _ in range(num_user)]
    
    user_id = (user_ids * (len(df_csv) // num_user))[:len(df_csv)]
    random.shuffle(user_id)
    df_csv['user_id'] = user_id
    
    return df_csv

test = generate_data_user(df_csv, 50)
df_csv.head()
```
